```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('train.csv')
sns.set(style="whitegrid")
```

```python
# 1. Dataset Overview
print("ðŸ”¹ Dataset Info:")
print(df.info())
print("\nðŸ”¹ Summary Statistics:")
print(df.describe())
```

```python
# 2. Value Counts for Categoricals
print("\nðŸ”¹ Value Counts:")
print("Survived:\n", df['Survived'].value_counts())
print("Pclass:\n", df['Pclass'].value_counts())
print("Sex:\n", df['Sex'].value_counts())
print("Embarked:\n", df['Embarked'].value_counts())
```

```python
# 3. Histograms
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")

plt.subplot(1, 2, 2)
sns.histplot(df['Fare'], bins=30, kde=True, color='lightgreen')
plt.title("Fare Distribution")
plt.tight_layout()
plt.show()
```

```python
# 4. Count Plots
plt.figure(figsize=(14, 5))t
plt.subplot(1, 2, 1)
sns.countplot(x='Survived', hue='Survived', data=df, palette='pastel', legend=False)
plt.title("Survival Count")

plt.subplot(1, 2, 2)
sns.countplot(x='Pclass', hue='Pclass', data=df, palette='muted', legend=False)
plt.title("Passenger Class Distribution")
plt.tight_layout()
plt.show()
```

```python
# 5. Boxplots: Age & Fare by Class
plt.figure(figsize=(14, 5))
# Age vs Pclass
plt.subplot(1, 2, 1)
sns.boxplot(x='Pclass', y='Age', hue='Pclass', data=df, palette='cool', legend=False)
plt.title("Age by Passenger Class")

# Fare vs Pclass
plt.subplot(1, 2, 2)
sns.boxplot(x='Pclass', y='Fare', hue='Pclass', data=df, palette='coolwarm', legend=False)
plt.title("Fare by Passenger Class")
plt.tight_layout()
plt.show()
```

```python
# 6. Heatmap: Correlation
plt.figure(figsize=(8, 6))
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 7. Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='husl')
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()
```
